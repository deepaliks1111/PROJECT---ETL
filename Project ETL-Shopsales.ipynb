```python
# TRANSFORMING AND CLEANING THE DATA
```

```python
combined = pd.concat([dfShopsales2,dfShopsales1])
combined
```

```python
comb_data = combined.dropna(axis = 0, how ='any')
```

```python
comb_data
```

```python
sort_data = comb_data.sort_values(["Outlet_Establishment_Year", "Item_Fat_Content"], axis=0, 
                 ascending=False) 
```

```python
sort_data
```

```python
clean_data = sort_data.drop(['Item_Identifier', 'Outlet_Identifier', 'Outlet_Location_Type', 'Outlet_Size', 'Item_Visibility', 'Item_Weight'], axis = 1)
clean_data
```

```python
comb_data = clean_data.rename(columns = {"Item_Fat_Content": "FAT CONTENT", 
                                  "Item_MRP":"MRP", 
                                  "Item_Outlet_Sales": "OUTLET SALES", "Item_Type": "ITEM TYPE", "Outlet_Establishment_Year":"OUTLET YEAR", "Outlet_Type":"OUTLET TYPE" }) 
```
